```python
# reading data in a dataframe
df = pd.read_csv("D:\Third Year\Machine Learning\Labs\Session1\salary.csv")
df.head()
```

```python
print('Missing Data\n')
print(df.isna())
print(df.isnull())
```

```python
# Print only missing data
print('Filter based on columns: \n')
df[df.isnull().any(axis=1)]
```

```python
# summing up the missing values
df.isnull().sum()
```

```python
# dropping the missing data
df2 = df.dropna()
df2
```

```python
# filling the missing data
df3 = df.copy()
df3['Salary'].fillna(df['Salary'].mean(), inplace=True)
df3

df['Salary'].fillna(df['Salary'].median(), inplace=True)
# we can fill the na , null values with the mean, median or whatever thing we feel correct and that resonates with the feel of the data
```

```python
df["Gender"].fillna(df["Gender"].mode()[0], inplace=True)
df["Country"].fillna(df["Country"].mode()[0], inplace=True)
df["Company"].fillna(df["Company"].mode()[0], inplace=True)
df
```
